# Data Visualisations

```python
data["region"].unique()
```

```python
region_df = data.groupby("region")["AveragePrice"].mean().to_frame().sort_values(by = "AveragePrice", ascending = False).head(10)
sns.barplot(region_df.index, region_df["AveragePrice"], data = region_df)
plt.title("Top 10 regions with best Average")
plt.xlabel("Regions")
plt.ylabel("Average Price")
plt.xticks(rotation = "vertical")
plt.show()
```

```python
region = data.groupby("region")["AveragePrice"].mean().to_frame()
plt.plot(region.index, region["AveragePrice"], label = "Plot line", marker = "o", lw = 4, ls = "-.", color = "#32a852", markeredgecolor = "#000000")
plt.xlabel("All the Regions")
plt.ylabel("Average Values")
plt.title("Average values of all the regions")
plt.xticks(rotation = "vertical")
plt.legend()
plt.show()
```

```python
data["year"].unique()
```

```python
sns.barplot(data["year"],data["AveragePrice"], hue = data["type"])
plt.xlabel("Years")
plt.ylabel("Average Price")
plt.legend()
plt.show()
```

```python
albany_df = data[data['region']=="Albany"]
albany_df.set_index("Date", inplace=True)
albany_df["AveragePrice"].plot()
plt.xlabel("Dates")
plt.ylabel("Average Price")
plt.title("Plot for Average Prices of Albany")
plt.show()
```

```python
albany_df["AveragePrice"].rolling(25).mean().plot()
plt.xlabel("Dates")
plt.ylabel("Average Price")
plt.title("Plot for Average Prices of Albany using rolling value")
plt.show()
```

# Data visualisation

```python
act_min_wage = pd.DataFrame()

for name, group in minwage.groupby("State"):
    if act_min_wage.empty:
        act_min_wage = group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name})
    else:
        act_min_wage = act_min_wage.join(group.set_index("Year")[["Low.2018"]].rename(columns={"Low.2018":name}))

act_min_wage.head()
```

```python
act_min_wage.corr()
```
